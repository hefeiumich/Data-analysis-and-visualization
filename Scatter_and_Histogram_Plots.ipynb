# Example codes for Scatter and Histogram Plots

Import python libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.metrics import mean_squared_error
from math import sqrt
%matplotlib inline
```

Generate example data

```python
N = 10000
y_test = np.random.randn(N,1) # the data is an example 
y_pred_test = np.random.randn(N,1) # the data is an example 
rms_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("root mean square error of test:", rms_test)
```

# Scatter Plot

```python
print("Scatter Plot")
ymin = min(min(y_test), min(y_pred_test))
ymax = max(max(y_test), max(y_pred_test))
# save the figure in pdf 
f = plt.figure()
plt.style.use('classic')
#print(plt.style.available)
best_fit = y_test
plt.plot(y_test, y_pred_test, 'o', label='RMSE = %3.2f m/s'  % rms_test)
plt.plot(y_test, best_fit, 'r--', label='One to one line', linewidth=2)

propdict =  {'fontsize': 12, 'fontweight': 'bold'}

font = {'weight' : 'bold',
        'size'   : 12}

plt.rc('font', **font)
plt.legend(loc='upper left')
plt.title('UTRACK', propdict)
plt.xlabel('Actual Wind (m/s)', propdict)
plt.ylabel('Predicted Wind (m/s)', propdict)
plt.xlim(ymin*0.90, ymax*1.03)
plt.ylim(ymin*0.90, ymax*1.03)
plt.rc('font', **font)
plt.show()
f.savefig("./pdf/scatter_plot_utrack_test_linear_best.pdf", bbox_inches='tight')
```

# Hist Plot

```python
print("Histogram Plot")
f = plt.figure()
plt.style.use('classic')
bins = np.linspace(ymin, ymax, 30)
print(bins)
plt.hist(y_pred_test, bins, alpha=0.5, label='Prediction')
plt.hist(y_test, bins, alpha=0.5, label='Actual')
plt.title('Histogram of Predicted and Actual Values', propdict)
plt.xlabel('Wind Speed (m/s)', propdict)
plt.ylabel('Frequency', propdict)
plt.legend(loc='upper right')
plt.show()
f.savefig("./pdf/hist_plot_utrack_test_linear_best.pdf", bbox_inches='tight')
```

# Prediction Error Plot

```python
print("Prediction Error Plot")
y_test_res = y_pred_test - y_test
f = plt.figure()
plt.style.use('classic')
plt.plot(y_test, y_test_res, 'o')
propdict =  {'fontsize': 12, 'fontweight': 'bold'}
font = {'weight' : 'bold',
        'size'   : 12}
plt.rc('font', **font)
plt.title('UTRACK', propdict)
plt.xlabel('Actual Wind (m/s)', propdict)
plt.ylabel('Prediction - Actual (m/s)', propdict)
plt.rc('font', **font)
plt.show()
f.savefig("./pdf/res_plot_utrack_test_linear_best.pdf", bbox_inches='tight')
```

# Histogram of the Prediction Error Plot

```python
f = plt.figure()
plt.style.use('classic')
# the histogram of the data 
sns.distplot(y_test_res, hist=True, kde=True, bins=20, color='blue', hist_kws={'edgecolor':'black'}, 
             kde_kws={'linewidth': 4, 'color': 'black'})
plt.xlabel('Prediction - Actual (m/s)', propdict)
plt.ylabel('Frequency', propdict)
plt.title('UTRACK', propdict)
plt.show()
f.savefig("./pdf/hist_res_utrack_test_linear_best.pdf", bbox_inches='tight')
```
